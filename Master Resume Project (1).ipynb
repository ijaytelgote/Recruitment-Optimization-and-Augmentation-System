```python
def recommend(resume,jobD):
  jobD =jobD[0]
  resumes=resume

  # Initialize the TF-IDF vectorizer
  tfidf_vectorizer = TfidfVectorizer(max_features=5000)

  # Fit and transform sentences
  tfidf_matrix = tfidf_vectorizer.fit_transform(resume)

  # Transform user input
  user_input_vector = tfidf_vectorizer.transform([jobD])

  # Calculate cosine similarities
  cosine_similarities = linear_kernel(user_input_vector, tfidf_matrix)

  # Get the most similar sentence indices
  similar_sentence_indices = cosine_similarities.argsort()[0][::-1]

  # Recommend the top 3 sentences
  top_recommendations = [resumes[i] for i in similar_sentence_indices[:2]]

  # Display recommendations
  print("Recommended Sentences:")
  for i, sentence in enumerate(top_recommendations):
      print (f"{i + 1}. {sentence}")
```

```python
##sentiment analysis
import spacy
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
def sent(text):
    doc = nlp(text)
    sentences = [sentence.text for sentence in doc.sents]

    for sentence in sentences:
        sentiment = analyzer.polarity_scores(sentence)
        return sentiment['compound']
```
